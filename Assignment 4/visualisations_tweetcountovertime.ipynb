```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("C:/Users/hussa/OneDrive - KU Leuven/Master 1/Semester 2/Advanced Analytics/Assignment 4/TweetCountOverTime.csv")
```

```python
# Convert 'TweetDate' column to datetime
df['TweetDate'] = pd.to_datetime(df['TweetDate'])

# Filter data starting from November 1, 2023
df = df[df['TweetDate'] >= '2023-11-01']

# Group by 'PosterParty' and aggregate 'TotalTweetCount' by sum
party_counts = df.groupby(['PosterParty', 'TweetDate']).agg({'TotalTweetCount': 'sum'}).reset_index()
party_counts_pivot = party_counts.pivot(index='TweetDate', columns='PosterParty', values='TotalTweetCount')

# Plotting
party_counts_pivot.plot.line(figsize=(10, 6))
plt.title('Total Tweet Counts Over Time by Party (Since November 2023)')
plt.xlabel('Date')
plt.ylabel('Total Tweet Count')
plt.grid(True)
plt.legend(title='Party')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
