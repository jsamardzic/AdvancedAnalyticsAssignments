# Preprocessing

```python
img_info.reviews.fillna(0, inplace=True)
img_info.percentage.fillna(-1, inplace=True)
```

# Keras

```python
import os

def get_size(path):
        size = os.path.getsize(path)
        if size < 1024:
            return f"{size} bytes"
        elif size < pow(1024,2):
            return f"{round(size/1024, 2)} KB"
        elif size < pow(1024,3):
            return f"{round(size/(pow(1024,2)), 2)} MB"
        elif size < pow(1024,4):
            return f"{round(size/(pow(1024,3)), 2)} GB"

get_size('/tmp')
```

## Load Data into Keras

```python
from itertools import chain

def create_label(x):
  labels = []
  for img_i in x.screenshots:
    labels.append(
        {img_i:{
           "reviews": x.reviews,
            'percentage': x.percentage,
    }}
        )

  return labels

labels = list(chain.from_iterable(img_info.apply(create_label, axis=1)))
```

```python
from keras.utils import image_dataset_from_directory
```

```python
image_dataset_from_directory(
    'images',
    labels=labels
)
```

# Train-Test Split

```python
from sklearn.model_selection import train_test_split
train, test = train_test_split(img_info, test_size=0.1, random_state=21)
```
